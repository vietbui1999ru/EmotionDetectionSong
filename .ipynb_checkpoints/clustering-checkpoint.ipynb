```python
df.count
```

```python
#Fit tranfsorm using minmax and kmeans. Categorize by transforming the song's features into a single interpretable digit
col_features = df.columns[6:]
X = MinMaxScaler().fit_transform(df[col_features])
```

```python
#since we only need to classify three moods (happy, sad, chill), the total clusters=3
kmeans = KMeans(init="k-means++",n_clusters=3,random_state=15).fit(X)
```

```python
#add new label at the end of the column section for kmean values
df['kmeans'] = kmeans.labels_
```

```python
#display new dataframe with updated kmeans label
df
```

Now that I think about it, I should have clustered before analyzing... 

```python
extreme_happy = df[(df['danceability'] > 0.57) & (df['valence'] > 0.38) & (df['loudness'] > -10)]
extreme_happy.sample(10)
```

```python
extreme_sad = df[(df['danceability'] < 0.3) & (df['valence'] < 0.2) & (df['loudness'] < -20)]
extreme_sad.sample(4)
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,projection='3d')

x = df['valence']
y = df['danceability']
z = df['loudness']
cmhot = cmhot = plt.get_cmap('Accent')
ax.scatter(x,y,z,c=df['kmeans'],s=40,cmap=cmhot)
ax.set_xlabel('Valence',fontsize=12)
ax.set_ylabel('Danceability',fontsize=12)
ax.set_zlabel('Loudness',fontsize=12)
ax.set_title("3D Scatter Plot of Songs Clustered")

```

We see that grey dots represent moody/sad songs, green for chill songs, and blue for happy and energetic songs

```python
df.groupby(['kmeans']).mean()
```

There's also a direct relationship between happy/energetic songs with their high popularity
and the lower popularity for sad songs.

```python
cluster_0 = df[df['kmeans']==0]
cluster_1  = df[df['kmeans']==1]
cluster_2 = df[df['kmeans']==2]
```

## ADDED: BEGIN CREATING NEW DATAFRAMES WITH NEW COLUMN: KMEANS (SAD, CHILL, HAPPY) ##

```python
cluster_chill = cluster_0.replace(to_replace=0, value='chill')
cluster_chill.reset_index(drop=True)
cluster_chill.index = np.arange(1, len(cluster_chill) + 1)
cluster_chill
```

```python
cluster_happy = cluster_1.replace(to_replace=1, value='happy')
cluster_happy.reset_index(drop=True)
cluster_happy.index = np.arange(1, len(cluster_happy) + 1)
cluster_happy
```

```python
cluster_sad = cluster_2.replace(to_replace=2, value='sad')
cluster_sad.reset_index(drop=True)
cluster_sad.index = np.arange(1, len(cluster_sad) + 1)
cluster_sad
```

```python
cluster_chill.to_csv("data/cluster_chill.csv",index=False)
cluster_happy.to_csv("data/cluster_happy.csv",index=False)
cluster_sad.to_csv("data/cluster_sad.csv",index=False)
df = pd.concat([cluster_happy, cluster_chill, cluster_sad])
df.to_csv("data/df_final.csv",index=False)
```
